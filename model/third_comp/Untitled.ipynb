Extract the month, hour and day from the variable date

```python
#train['calldate'] = train['calldate'].apply(lambda x: datetime.strftime(x,"%Y-%m-%d-%H"))
```

```python
train['year']=train['calldate'].apply(lambda x: datetime.strftime(x,"%Y"))
train['month']=train['calldate'].apply(lambda x: datetime.strftime(x,"%m"))
train['day']=train['calldate'].apply(lambda x: datetime.strftime(x,"%d"))
train['hour']=train['calldate'].apply(lambda x: datetime.strftime(x,"%H"))
```

```python
train.head(5)
```

Now we need to count the number of call per hour per day

```python
train['num_of_call']=1
data= (train.groupby(['month',"day", 'hour'])['num_of_call'].sum().reset_index())
train['num_of_call']=data['num_of_call']
```
